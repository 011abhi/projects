```python
def match_patterns(img_array,pattern,number):
    email=[]
    phno=[]
    dict1=pytesseract.image_to_data(img_array,output_type=pytesseract.Output.DICT)
    for i in range(len(dict1["left"])):
        if dict1["conf"][i]>30:
            text=dict1["text"][i]            
            if re.match(pattern,text) :
                email.append(text)
                x,y,w,h=dict1["left"][i],dict1["top"][i],dict1["width"][i],dict1["height"][i]
                cv2.rectangle(img_array,(x,y),(x+w,y+h),(0,25,255),2)
        if dict1["text"][i].isnumeric(): 
            num=dict1["text"][i]
            if re.fullmatch(number,num) :
                phno.append(num)
                x,y,w,h=dict1["left"][i],dict1["top"][i],dict1["width"][i],dict1["height"][i]
                cv2.rectangle(img_array,(x,y),(x+w,y+h),(0,25,255),2)

    cv2.imshow("ocr",img_array)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    return email,phno
```

```python
text=cv2.imread("idcard.jpeg")
pattern=r"[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.com"
number=r"\+\d{3}-\d{3}-\d{4}"
match_patterns(text,pattern,number)
```
